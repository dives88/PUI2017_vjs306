**c. suggested test for H0**

We need to test whether two means differ:
* mean age of males
* mean age of females

More specifically, we are testing whether $M_m >= F_m$. 

Therefore we need a test which is:
 - comparing two means against each other
 - a one-tailed test.
 
The proposed test is a two-sample t-test. We meet the relevant conditions for using this test, since the two samples are randomly distributed from the larger population of active citibike members (they were selected based on the one-month period), and the sample size is large.

The t test is calculated as follows:

$t= (\bar{x_1} - \bar{x_2}) - (\mu_1 - \mu_2) / {\sqrt{s^2_p(\frac{1}{n_1}+\frac{1}{n_2})}}$

where $\bar{x_1}$ and $\bar{x_2}$ are the sample means;
$\mu_1$ and $\mu_2$ are the population means;
$n_1$ and $n_2$ are the sample sizes;
{s^2_p$ is the pooled variance; and
$s^2_1$ and $s^2_2$ are the sample variances.

The study can use a significance level: $\alpha$ = 0.05.

We need to calculate the degrees of freedom for the study, which will be $n_1 + n_2 - 2$. Then we use a t table to look up the critical value for $\alpha=0.05, df = (n_1 + n_2 -2)$. We call this $v_crit$. Note we are using the t statistic for a one-sided test.

Then we calculate the test statistic, $t$.

Decision rule:
We reject H0 if $t > v_crit$.
